```python
#video.ittensive.com/data/018-python-advanced/beru.ru/
import requests
import pandas as pd
from bs4 import BeautifulSoup
headers = {"User-Agent": "ittensive-python-Cheremin/1.0 (+https://www.ittensive.com)"}
r = requests.get("http://video.ittensive.com/data/018-python-advanced/beru.ru/",headers=headers) #загружаем HTMLкод таблицы
html = BeautifulSoup(r.content) #отправляем в SOUP
links = html.find_all("a", {"class": "_3ioN70chUh Usp3kX1MNT _3Uc73lzxcf"}) #находим класс с описанием товаров
link_263 = ''
link_452 = ''
for link in links:
    if str(link).find("Саратов 263") > -1:# находим в теге<а /а> текс
        link_263 = link["href"] # берём из класса href текст с ссылкой
    if str(link).find("Саратов 452") > -1:  
        link_452 = link["href"]
def find_volume (link):
    r = requests.get("http://video.ittensive.com/data/018-python-advanced/beru.ru/" + link) 
    html = BeautifulSoup(r.content)
    volume = html.find_all("span", {"class": "_112Tad-7AP"})
    return int(''.join(i for i in volume[2].get_text() if i.isdigit()))

if link_263 and link_452:
    volume_263 = find_volume(link_263)
    volume_452 = find_volume(link_452)
    diff = max(volume_263, volume_452) - min(volume_263, volume_452)
    print (diff)


```
