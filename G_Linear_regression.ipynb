```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import scipy.stats as st
ch = pd.read_csv("http://oblakoural.ru/demo13h.csv")  
print(ch)
plt.scatter(ch["year"], ch["size"])
plt.show()
```

```python
x = np.array(ch["year"]).reshape(-1, 1)
y = np.array(ch["size"]).reshape(-1, 1)
#print (x)
#print (y)
model = LinearRegression().fit(x, y)
```

```python
r_sq = model.score(x, y)
y_ = np.array(x * model.coef_ + model.intercept_)
print (model.coef_)
print (model.intercept_)
print (r_sq)
```

```python
yy_ = np.array(y - y_).flatten() # ошибка регрессии
print(yy_)
shapiro = st.shapiro(yy_) # Тест ( Тест Шапиро-Уилка для распределения ошибок
print (shapiro [1])
```
